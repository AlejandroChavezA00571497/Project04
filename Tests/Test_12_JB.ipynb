```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib as plt
import sklearn as skl
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
import numpy as np
```

## Importing data

```python
df = pd.read_csv("../Resources/attributesBillboard.csv")
df.drop(columns=["Unnamed: 0"], inplace=True)
df.head()
```

```python
df2 = pd.read_csv("../Resources/songAttributes_1999-2019.csv", encoding='unicode_escape')
df2.drop(columns=["X"], inplace=True)
df2["Popularity"].describe()
df2
```

```python
df2['Popularity'].describe()
```

```python
df['Popularity'] = np.where((df['Popularity'] > 58) & df['Popularity']
                                            .notna(), 1, 0)
```

```python
df2['Popularity'] = np.where((df2['Popularity'] > 31) & df2['Popularity']
                                            .notna(), 1, 0)
```

# Attributes Billboard

```python
X = df[['Valence', 'Energy', "Danceability", "Acousticness", "Tempo", "Duration", "Instrumentalness","Liveness","Loudness"]]
y = df['Popularity']
```

```python
df["Popularity"].value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
nn_model = tf.keras.models.Sequential()

nn_model.add(tf.keras.layers.Dense(units=20, activation="relu", input_dim=9))

nn_model.add(tf.keras.layers.Dense(units=20, activation="relu"))

nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the Sequential model
nn_model.summary()
```

```python
# Compile the model and train over more than 100 epochs
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

fit_model = nn_model.fit(X_train_scaled, y_train, epochs=200)
```

```python
# Evaluate the performance of model using the loss and predictive accuracy of the model on the test dataset.
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
