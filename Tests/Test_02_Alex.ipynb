```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv("../Resources/songAttributes_1999-2019.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns= ["Name", "Unnamed: 0", "Album", "Artist"])
```

```python
df = pd.get_dummies(df)
df.info()
```

```python
df["is_popular"] = (df["Popularity"] >= 40)
df
```

```python
df_02 = df[["Explicit", "Instrumentalness","Mode","is_popular"]]
df_02
```

```python
# Split our preprocessed data into our features and target arrays
X = df.drop(columns = ["is_popular"])
y = df["is_popular"]
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input_features = len(X_train_scaled[0])
num_neurons_hl1 = 5
num_neurons_hl2 = 5


nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units = num_neurons_hl1, input_dim = num_input_features, activation = "relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units = num_neurons_hl2, activation = "relu"))


# Output layer
nn.add(tf.keras.layers.Dense(units = 1, activation = "sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
fit_model = nn.fit(X_train_scaled, y_train, epochs = 5)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
