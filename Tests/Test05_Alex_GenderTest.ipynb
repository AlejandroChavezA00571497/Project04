```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("../Resources/artistDf.csv")
df = df.dropna()
df.info()
```

```python
df
```

```python
df = df.drop(columns= ["Artist", "Genres", "X"])
df
```

```python
df_dummies = pd.get_dummies(df)
df_dummies
```

```python
X = df_dummies.drop(columns = ["Followers"])
y = df_dummies["Followers"]

X = X.astype(np.float32)
y = y.astype(np.float32)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model
model = Sequential()
model.add(Dense(64, input_dim=X.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))  # Output layer with one neuron for regression

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, validation_split=0.2)
```
